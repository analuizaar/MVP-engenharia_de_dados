# Pergunta 4 -	Qual é o volume de pendências criadas e sua urgência?

```python
%sql
SELECT COUNT(*) AS total_pendencias
FROM gold.fato_pendencia_criada;
```

% das visitas com pendencias

```python
%sql
SELECT 
    ROUND(
        100.0 * COUNT(DISTINCT p.visita_id)
        / COUNT(DISTINCT v.visita_id),
        2
    ) AS percentual_visitas_com_pendencias
FROM gold.dim_visita v
LEFT JOIN gold.fato_pendencia_criada p
    ON v.visita_id = p.visita_id;
```

divisao da urgencia das pendencias. urgente = true

```python
%sql
SELECT 
    eh_urgente,
    COUNT(*) AS total
FROM gold.fato_pendencia_criada
GROUP BY eh_urgente
ORDER BY eh_urgente DESC;
```

```python
%sql
SELECT 
    p.nome_projeto,
    COUNT(*) AS total_pendencias
FROM gold.fato_pendencia_criada fp
JOIN gold.dim_projeto p ON fp.projeto_id = p.projeto_id
GROUP BY p.nome_projeto
ORDER BY total_pendencias DESC;
```

```python
%sql
SELECT 
    u.email,
    COUNT(*) AS pendencias_criadas
FROM gold.fato_pendencia_criada fp
JOIN gold.dim_usuario u ON fp.usuario_id = u.usuario_id
GROUP BY u.email
ORDER BY pendencias_criadas DESC
LIMIT 20;
```

pendencias resolvidas (true) ou nao (false)

```python
%sql
SELECT 
    resolvida,
    COUNT(*) AS total
FROM gold.fato_pendencia_criada
GROUP BY resolvida;
```

pendencia por dia da semana

```python
%sql
SELECT 
    t.dia_semana,
    COUNT(*) AS total
FROM gold.fato_pendencia_criada fp
JOIN gold.dim_tempo t ON fp.tempo_id = t.tempo_id
GROUP BY t.dia_semana
ORDER BY total DESC;
```
