# Pergunta 2: Quantos projetos e visitas são criados por usuário?

### HISTOGRAMA: Número de PROJETOS por USUÁRIO

```python
%sql
WITH projetos_por_usuario AS (
    SELECT 
        u.usuario_id,
        COUNT(fv.projeto_id) AS qtd_projetos
    FROM gold.dim_usuario u
    LEFT JOIN gold.fato_visita_criada fv 
        ON fv.usuario_id = u.usuario_id
    GROUP BY u.usuario_id
)

SELECT
    CASE 
        WHEN qtd_projetos = 0 THEN 'Nenhum projeto'
        WHEN qtd_projetos = 1 THEN '1 projeto'
        WHEN qtd_projetos BETWEEN 2 AND 4 THEN '2 a 4 projetos'
        WHEN qtd_projetos BETWEEN 5 AND 10 THEN '5 a 10 projetos'
        WHEN qtd_projetos > 10 THEN '10+ projetos'
    END AS faixa,
    COUNT(*) AS usuarios_nessa_faixa
FROM projetos_por_usuario
GROUP BY faixa
ORDER BY faixa;
```

```python
%sql
WITH visitas_por_projeto AS (
    SELECT 
        p.projeto_id,
        COUNT(f.id_fato_visita) AS qtd_visitas
    FROM gold.dim_projeto p
    LEFT JOIN gold.fato_visita_criada f
        ON f.projeto_id = p.projeto_id
    GROUP BY p.projeto_id
)

SELECT
    CASE 
        WHEN qtd_visitas = 0 THEN 'a) 0 visitas'
        WHEN qtd_visitas = 1 THEN 'b) 1 visita'
        WHEN qtd_visitas BETWEEN 2 AND 3 THEN 'c) 2 a 3 visitas'
        WHEN qtd_visitas BETWEEN 4 AND 5 THEN 'd) 4 a 5 visitas'
        WHEN qtd_visitas BETWEEN 6 AND 10 THEN 'e) 6 a 10 visitas'
        WHEN qtd_visitas BETWEEN 11 AND 20 THEN 'f) 11 a 20 visitas'
        WHEN qtd_visitas > 20 THEN 'g) 20+ visitas'
    END AS faixa,
    COUNT(*) AS projetos_nessa_faixa
FROM visitas_por_projeto
GROUP BY faixa
ORDER BY faixa;

```

usuarios que mais criaram projetos e visitas

```python
%sql
SELECT 
    u.email,
    COUNT(DISTINCT fv.projeto_id) AS projetos_criados
FROM gold.fato_visita_criada fv
JOIN gold.dim_usuario u ON fv.usuario_id = u.usuario_id
GROUP BY u.email
ORDER BY projetos_criados DESC
LIMIT 10;
```

```python
%sql
SELECT 
    p.nome_projeto,
    COUNT(*) AS visitas
FROM gold.fato_visita_criada fv
JOIN gold.dim_projeto p ON fv.projeto_id = p.projeto_id
GROUP BY p.nome_projeto
ORDER BY visitas DESC
LIMIT 10;

```

```python
%sql
WITH visitas_por_usuario AS (
    SELECT 
        usuario_id,
        COUNT(*) AS visitas
    FROM gold.fato_visita_criada
    GROUP BY usuario_id
)
SELECT
    CASE 
        WHEN visitas BETWEEN 0 AND 5 THEN '0 a 5 visitas'
        WHEN visitas BETWEEN 6 AND 20 THEN '6 a 20 visitas'
        WHEN visitas BETWEEN 21 AND 50 THEN '21 a 50 visitas'
        WHEN visitas > 50 THEN '50+ visitas'
    END AS faixa,
    COUNT(*) AS usuarios_nessa_faixa
FROM visitas_por_usuario
GROUP BY faixa
ORDER BY faixa;
```

```python
%sql
SELECT 
    p.nome_projeto,
    u.email,
    COUNT(*) AS visitas_do_usuario_no_projeto
FROM gold.fato_visita_criada fv
JOIN gold.dim_usuario u ON fv.usuario_id = u.usuario_id
JOIN gold.dim_projeto p ON fv.projeto_id = p.projeto_id
GROUP BY p.nome_projeto, u.email
ORDER BY visitas_do_usuario_no_projeto DESC
LIMIT 20;
```

quantas visitas cada projeto tem em media, e o maximo de visitas em um projeto

```python
%sql
WITH visitas AS (
    SELECT projeto_id, COUNT(*) AS qtd FROM gold.fato_visita_criada GROUP BY projeto_id
)
SELECT 
    MIN(qtd) AS min_visitas,
    AVG(qtd) AS avg_visitas,
    MAX(qtd) AS max_visitas
FROM visitas;
```
