# Pergunta 1 - Número de usuários ativos e frequência de uso

## Contar os usuários ativos (pelo menos 1 foto ou 1 pendência criada) e me dar a lista dos dias da semana com mais uso

```python
%sql
SELECT COUNT(DISTINCT usuario_id) AS usuarios_ativos
FROM (
    SELECT usuario_id FROM gold.fato_foto_adicionada
    UNION
    SELECT usuario_id FROM gold.fato_pendencia_criada
) x;
```

```python
%sql
SELECT COUNT (DISTINCT usuario_id)
FROM gold.dim_usuario;
```

Os usuarios mais ativos

```python
%sql
SELECT 
    t.dia_semana,
    COUNT(*) AS visitas
FROM gold.fato_visita_criada fv
JOIN gold.dim_tempo t ON fv.tempo_id = t.tempo_id
GROUP BY t.dia_semana
ORDER BY visitas DESC;
```

dia que as pessoas mais adicionam foto

```python
%sql
SELECT 
    t.dia_semana,
    COUNT(*) AS visitas
FROM gold.fato_foto_adicionada fv
JOIN gold.dim_tempo t ON fv.tempo_id = t.tempo_id
GROUP BY t.dia_semana
ORDER BY visitas DESC;
```

## Quais foram os usuarios que usaram o gelato em mais dias

```python
%sql
SELECT 
    u.email,
    COUNT(DISTINCT t.data) AS dias_ativos
FROM gold.fato_visita_criada fv
JOIN gold.dim_usuario u ON fv.usuario_id = u.usuario_id
JOIN gold.dim_tempo t ON fv.tempo_id = t.tempo_id
GROUP BY u.email
ORDER BY dias_ativos DESC
limit 20;
```
