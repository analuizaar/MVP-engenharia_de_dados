# Pergunta 5 - Qual o uso da funcionalidade anotação dentro das visitas?

quantas visitas tem anotações preenchidas

```python
%sql
SELECT 
    COUNT(*) AS visitas_com_anotacao
FROM gold.dim_visita
WHERE anotacoes IS NOT NULL AND trim(anotacoes) <> '';
```

proporção das visitas com anotações

```python
%sql
SELECT 
    ROUND(
        100.0 * COUNT(*) FILTER (WHERE anotacoes IS NOT NULL AND trim(anotacoes) <> '') 
        / COUNT(*),
        2
    ) AS percentual_com_anotacao
FROM gold.dim_visita;
```

quais usuarios mais usaram a feature anotação

```python
%sql
SELECT 
    u.email,
    COUNT(*) AS qtd_anotacoes
FROM gold.fato_visita_criada f
JOIN gold.dim_usuario u 
    ON f.usuario_id = u.usuario_id
JOIN gold.dim_visita v 
    ON f.nome_visita = v.nome_visita
WHERE v.anotacoes IS NOT NULL 
  AND trim(v.anotacoes) <> ''
GROUP BY u.email
ORDER BY qtd_anotacoes DESC;

```

histograma do tamanho das anotações

```python
%sql
WITH anot AS (
    SELECT 
        visita_id,
        LENGTH(anotacoes) AS tamanho
    FROM gold.dim_visita
    WHERE anotacoes IS NOT NULL AND trim(anotacoes) <> ''
)

SELECT 
    CASE 
        WHEN tamanho BETWEEN 1 AND 50 THEN '1–50 caracteres'
        WHEN tamanho BETWEEN 51 AND 150 THEN '51–150 caracteres'
        WHEN tamanho BETWEEN 151 AND 300 THEN '151–300 caracteres'
        WHEN tamanho BETWEEN 301 AND 600 THEN '301–600 caracteres'
        ELSE '600+ caracteres'
    END AS faixa,
    COUNT(*) AS qtd_visitas
FROM anot
GROUP BY faixa
ORDER BY faixa;
```

lista das principais anotações

```python
%sql
SELECT 
    nome_visita,
    anotacoes
FROM gold.dim_visita
WHERE anotacoes IS NOT NULL AND trim(anotacoes) <> ''
ORDER BY visita_id
LIMIT 20;

```
